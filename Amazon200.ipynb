```python
import csv
import requests
from bs4 import BeautifulSoup
import time

# def get_product_description(url):
#     try:
#         headers = {
#             'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36'
#         }
#         response = requests.get(url, headers=headers)
#         soup = BeautifulSoup(response.content, 'html.parser')
#         description_elements = [
#             soup.find('div', {'id': 'feature-bullets'}),
#             soup.find('div', {'id': 'productDescription'}),
#             soup.find('div', {'id': 'dpx-amazon-technical-details'}),
#         ]

#         for element in description_elements:
#             if element:
#                 product_description = element.text.strip()
#                 break
#         else:
#             product_description = 'Description not available'
#     except Exception as e:
#         print(f"Error fetching description for {url}: {e}")
#         product_description = 'Description not available'

#     return product_description

def get_product_description(url):
    try:
        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36'
        }
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.content, 'html.parser')
        description_element = soup.find('div', {'id': 'feature-bullets'})
        #description_element2 = soup.find('div', {'class': 'a-list-item'})

        # description_elements = [
        #     soup.find('div', {'id': 'feature-bullets'}),
        #     soup.find('div', {'id': 'productDescription'}),
        #     soup.find('div', {'id': 'dpx-amazon-technical-details'}),
        # ]

        product_description = description_element.text.strip() if description_element else 'Description not available'
    except Exception as e:
        print(f"Error fetching description for {url}: {e}")
        product_description = 'Description not available'

    return product_description

def get_product_data(keyword, num_products):
    base_url = f'https://www.amazon.in/s?k={keyword}&ref=nb_sb_noss_1'
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36'
    }
    product_data = []

    for page in range(1, (num_products // 48) + 2):
        url = f'{base_url}&page={page}'
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.content, 'html.parser')

        product_containers = soup.find_all('div', {'data-component-type': 's-search-result'})

        for container in product_containers:
            product_title = container.find('h2').text.strip()
            product_price = container.find('span', {'class': 'a-offscreen'})
            product_price = product_price.text.strip() if product_price else 'Not available'
            product_rating = container.find('span', {'class': 'a-icon-alt'})
            product_rating = product_rating.text.split()[0] if product_rating else 'Not available'
            product_reviews = container.find('span', {'class': 'a-size-base'})
            product_reviews = product_reviews.text.strip() if product_reviews else 'Not available'
            product_url = container.find('a', {'class': 'a-link-normal'})['href']
            product_url = f'https://www.amazon.in{product_url}' if product_url else 'Not available'
            product_description = get_product_description(product_url)

            product_data.append([product_title, product_price, product_rating, product_reviews, product_url, product_description])
        
        # Introduce a delay to avoid overloading the server
        time.sleep(2)

    return product_data

def save_to_csv(data, filename):
    with open(filename, mode='w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(['Product Title', 'Price', 'Rating', 'Reviews', 'Product URL', 'Product Description'])
        writer.writerows(data)

def display_csv(filename):
    with open(filename, mode='r', newline='', encoding='utf-8') as file:
        reader = csv.reader(file)
        for row in reader:
            print(', '.join(row))

if __name__ == "__main__":
    keyword = 'bag'  # Replace 'bag' with your desired keyword
    num_products = 2000
    filename = r'C:\Users\sivaa\OneDrive\Documents\Desktop\Python Default Task\amazon_final.csv'

    product_data = get_product_data(keyword, num_products)
    save_to_csv(product_data, filename)

    print(f'Product data saved to "{filename}".')

    # Display the CSV contents
    print("\nDisplaying CSV Contents:")
    display_csv(filename)
```
